```python
tas_binaire = []
```

```python
def ajoute(tas, elt):
    """Ajoute un élément dans le tas binaire."""
    tas.append(elt)
    i = len(tas) - 1
    while i > 0:
        parent = (i - 1) // 2
        if tas[i] < tas[parent]:
            tas[i], tas[parent] = tas[parent], tas[i]
            i = parent
        else:
            break

def enleve(tas):
    """Enlève l'élément le plus petit du tas binaire."""
    if len(tas) == 0:
        return None
    if len(tas) == 1:
        return tas.pop()
    
    racine = tas[0]
    tas[0] = tas.pop()
    
    i = 0
    while True:
        gauche = 2 * i + 1
        droite = 2 * i + 2
        plus_petit = i
        
        if gauche < len(tas) and tas[gauche] < tas[plus_petit]:
            plus_petit = gauche
        if droite < len(tas) and tas[droite] < tas[plus_petit]:
            plus_petit = droite
        
        if plus_petit == i:
            break
        
        tas[i], tas[plus_petit] = tas[plus_petit], tas[i]
        i = plus_petit
    
    return racine
```
